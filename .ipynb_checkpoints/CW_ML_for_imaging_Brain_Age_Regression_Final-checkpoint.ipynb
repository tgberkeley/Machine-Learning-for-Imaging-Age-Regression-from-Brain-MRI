### Set up a simple medical image viewer and import SimpleITK

```python
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt

from ipywidgets import interact, fixed
from IPython.display import display

# Calculate parameters low and high from window and level
def wl_to_lh(window, level):
    low = level - window/2
    high = level + window/2
    return low,high

def display_image(img, x=None, y=None, z=None, window=None, level=None, colormap='gray', crosshair=False):
    # Convert SimpleITK image to NumPy array
    img_array = sitk.GetArrayFromImage(img)
    
    # Get image dimensions in millimetres
    size = img.GetSize()
    spacing = img.GetSpacing()
    width  = size[0] * spacing[0]
    height = size[1] * spacing[1]
    depth  = size[2] * spacing[2]
    
    if x is None:
        x = np.floor(size[0]/2).astype(int)
    if y is None:
        y = np.floor(size[1]/2).astype(int)
    if z is None:
        z = np.floor(size[2]/2).astype(int)
    
    if window is None:
        window = np.max(img_array) - np.min(img_array)
    
    if level is None:
        level = window / 2 + np.min(img_array)
    
    low,high = wl_to_lh(window,level)

    # Display the orthogonal slices
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 4))

    ax1.imshow(img_array[z,:,:], cmap=colormap, clim=(low, high), extent=(0, width, height, 0))
    ax2.imshow(img_array[:,y,:], origin='lower', cmap=colormap, clim=(low, high), extent=(0, width,  0, depth))
    ax3.imshow(img_array[:,:,x], origin='lower', cmap=colormap, clim=(low, high), extent=(0, height, 0, depth))

    # Additionally display crosshairs
    if crosshair:
        ax1.axhline(y * spacing[1], lw=1)
        ax1.axvline(x * spacing[0], lw=1)
        ax2.axhline(z * spacing[2], lw=1)
        ax2.axvline(x * spacing[0], lw=1)
        ax3.axhline(z * spacing[2], lw=1)
        ax3.axvline(y * spacing[1], lw=1)

    plt.show()
    
def interactive_view(img):
    size = img.GetSize() 
    img_array = sitk.GetArrayFromImage(img)
    interact(display_image,img=fixed(img),
             x=(0, size[0] - 1),
             y=(0, size[1] - 1),
             z=(0, size[2] - 1),
             window=(0,np.max(img_array) - np.min(img_array)),
             level=(np.min(img_array),np.max(img_array)));
```

## Part A: Age prediction using brain tissue segmentation and volume features

The first approach aims to regress the age of a subject using the volumes of brain tissues as features. The brain tissues include grey matter (GM), white matter (WM), and cerebrospinal fluid (CSF). It is known that with increasing age the ventricles enlarge (filled with CSF), while it is assumed that grey and white matter volume may decrease over time. However, as overall brain volume varies across individuals, taking the absolute volumes of tissues might not be predictive. Instead, relative volumes need to be computed as the ratios between each tissue volume and overall brain volume. To this end, a four-class (GM, WM, CSF, and background) brain segmentation needs to be implemented which will be trained using a total of 52 subjects (47 for training, 5 for validation). The segmentation method is then applied to the remaining 600 brain scans which will be used to train and test the age regression. #### Data Helpers

```python
def zero_mean_unit_var(image, mask):
    """Normalizes an image to zero mean and unit variance."""

    img_array = sitk.GetArrayFromImage(image)
    img_array = img_array.astype(np.float32)

    msk_array = sitk.GetArrayFromImage(mask)

    mean = np.mean(img_array[msk_array>0])
    std = np.std(img_array[msk_array>0])

    if std > 0:
        img_array = (img_array - mean) / std
        img_array[msk_array==0] = 0

    image_normalised = sitk.GetImageFromArray(img_array)
    image_normalised.CopyInformation(image)

    return image_normalised


def resample_image(image, out_spacing=(1.0, 1.0, 1.0), out_size=None, is_label=False, pad_value=0):
    """Resamples an image to given element spacing and output size."""

    original_spacing = np.array(image.GetSpacing())
    original_size = np.array(image.GetSize())

    if out_size is None:
        out_size = np.round(np.array(original_size * original_spacing / np.array(out_spacing))).astype(int)
    else:
        out_size = np.array(out_size)

    original_direction = np.array(image.GetDirection()).reshape(len(original_spacing),-1)
    original_center = (np.array(original_size, dtype=float) - 1.0) / 2.0 * original_spacing
    out_center = (np.array(out_size, dtype=float) - 1.0) / 2.0 * np.array(out_spacing)

    original_center = np.matmul(original_direction, original_center)
    out_center = np.matmul(original_direction, out_center)
    out_origin = np.array(image.GetOrigin()) + (original_center - out_center)

    resample = sitk.ResampleImageFilter()
    resample.SetOutputSpacing(out_spacing)
    resample.SetSize(out_size.tolist())
    resample.SetOutputDirection(image.GetDirection())
    resample.SetOutputOrigin(out_origin.tolist())
    resample.SetTransform(sitk.Transform())
    resample.SetDefaultPixelValue(pad_value)

    if is_label:
        resample.SetInterpolator(sitk.sitkNearestNeighbor)
    else:
        resample.SetInterpolator(sitk.sitkBSpline)

    return resample.Execute(image)


class ImageSegmentationDataset(Dataset):
    """Dataset for image segmentation."""

    def __init__(self, file_list_img, file_list_seg, file_list_msk, img_spacing, img_size):
        self.samples = []
        self.img_names = []
        self.seg_names = []
        for idx, _ in enumerate(tqdm(range(len(file_list_img)), desc='Loading Data')):
            img_path = file_list_img[idx]
            seg_path = file_list_seg[idx]
            msk_path = file_list_msk[idx]

            img = sitk.ReadImage(img_path, sitk.sitkFloat32)

            seg = sitk.ReadImage(seg_path, sitk.sitkInt64)

            msk = sitk.ReadImage(msk_path, sitk.sitkUInt8)

            #pre=processing
            img = zero_mean_unit_var(img, msk)
            img = resample_image(img, img_spacing, img_size, is_label=False)
            seg = resample_image(seg, img_spacing, img_size, is_label=True)
            msk = resample_image(msk, img_spacing, img_size, is_label=True)

            sample = {'img': img, 'seg': seg, 'msk': msk}

            self.samples.append(sample)
            self.img_names.append(os.path.basename(img_path))
            self.seg_names.append(os.path.basename(seg_path))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, item):
        sample = self.samples[item]

        image = torch.from_numpy(sitk.GetArrayFromImage(sample['img'])).unsqueeze(0)
        seg = torch.from_numpy(sitk.GetArrayFromImage(sample['seg'])).unsqueeze(0)
        msk = torch.from_numpy(sitk.GetArrayFromImage(sample['msk'])).unsqueeze(0)

        return {'img': image, 'seg': seg, 'msk': msk}

    def get_sample(self, item):
        return self.samples[item]

    def get_img_name(self, item):
        return self.img_names[item]

    def get_seg_name(self, item):
        return self.seg_names[item]
```

#### The Model

**TASK:** This is the **key part of task A-1** where you have to design a suitable CNN model for brain segmentation. The simple model provided below works to some degree (it let's you run through the upcoming cells), but it will not perform very well. Use what you learned in the lectures to come up with a good architecture. Start with a simple, shallow model and only increase complexity (e.g., number of layers) if needed.

```python
########################################
# BUILD A BETTER MODEL HERE
########################################

class SimpleNet3D(nn.Module):

    def __init__(self, num_classes):
        super(SimpleNet3D, self).__init__()
        self.conv1 = nn.Conv3d(1, 4, kernel_size=3, padding=1)
        self.conv2 = nn.Conv3d(4, num_classes, kernel_size=3, padding=1)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.conv2(x)
        
        return x # cross-entropy loss expects raw logits and applies softmax


########################################
# THE FRENCH MONSTER
########################################

class FrenchMonster(nn.Module):

    def __init__(self, num_classes):
        super(FrenchMonster, self).__init__()

        # reduce image dimensions
        self.main1 = nn.Sequential(
          nn.Conv3d(1, 16, kernel_size=3, stride=1, padding=1),
          nn.ReLU(inplace=True),
          nn.Conv3d(16, 16, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(16),
          nn.ReLU(inplace=True)
          )
        self.main2 = nn.Sequential(
          nn.Conv3d(16, 32, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(32),
          nn.ReLU(inplace=True),
          nn.Conv3d(32, 32, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(32),
          nn.ReLU(inplace=True)
          )
        self.main3 = nn.Sequential(
          nn.Conv3d(32, 64, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(64),
          nn.ReLU(inplace=True),
          nn.Conv3d(64, 64, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(64),
          nn.ReLU(inplace=True)
          )  
        
        # bottle neck
        self.main = nn.Sequential(
          nn.Conv3d(64, 128, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(128),
          nn.ReLU(inplace=True),
          nn.Conv3d(128, 128, kernel_size=3, stride=1, padding=1),
          nn.BatchNorm3d(128),
          nn.ReLU(inplace=True)
        )

        # increase image dimensions
        self.main4_1 = nn.Sequential(
            nn.ConvTranspose3d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm3d(64))
        self.main4_2 = nn.Sequential(
            nn.Conv3d(128, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm3d(64),
            nn.ReLU(inplace=True),
            nn.Conv3d(64, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm3d(64),
            nn.ReLU(inplace=True))
        
        self.main5_1 = nn.Sequential(
            nn.ConvTranspose3d(64, 32, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm3d(32))
        self.main5_2 = nn.Sequential(
            nn.Conv3d(64, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm3d(32),
            nn.ReLU(inplace=True),
            nn.Conv3d(32, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm3d(32),
            nn.ReLU(inplace=True))
        
        self.main6_1 = nn.Sequential(
            nn.ConvTranspose3d(32, 16, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm3d(16))
        self.main6_2 = nn.Sequential(
            nn.Conv3d(32, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm3d(16),
            nn.ReLU(inplace=True),
            nn.Conv3d(16, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm3d(16),
            nn.ReLU(inplace=True))

        # reduce the number of feature maps
        self.main7 = nn.Sequential(
            nn.Conv3d(16, num_classes, kernel_size=1, stride=1, padding=0)
        )

    def forward(self, x):

        # reduce image dimensions
        fm1 = self.main1(x)
        x = F.max_pool3d(fm1, kernel_size=2, stride=2)

        fm2 = self.main2(x)
        x = F.max_pool3d(fm2, kernel_size=2, stride=2)

        fm3 = self.main3(x)
        x = F.max_pool3d(fm3, kernel_size=2, stride=2)

        # bottle neck
        x = self.main(x)

        # increase image dimensions
        x = self.main4_1(x)
        x = torch.cat([x,fm3], dim=1)
        x = self.main4_2(x)

        x = self.main5_1(x)
        x = torch.cat([x, fm2], dim=1)
        x = self.main5_2(x)

        x = self.main6_1(x)
        x = torch.cat([x, fm1], dim=1)
        x = self.main6_2(x)

        # reduce the number of feature maps
        x = self.main7(x)

        
        return x # cross-entropy loss expects raw logits and applies softmax
```

```python
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)
```

#### TRAINING

Below is an implementation of a full training procedure including a loop for intermediate evaluation of the model on the validation data. Feel free to modify this procedure. For example, in addition to the loss you may want to monitor Dice scores (or others).

```python
from sklearn.metrics import f1_score

model_dir = os.path.join(out_dir, 'model')
if not os.path.exists(model_dir):
    os.makedirs(model_dir)

torch.manual_seed(rnd_seed) #fix random seed

# model = SimpleNet3D(num_classes=num_classes).to(device)
model = FrenchMonster(num_classes=num_classes).to(device)
model.apply(weights_init)
params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("Total number of parameters in the FrenchMonster: {}".format(params))
print('\n')
model.train()
    
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

loss_train_log = []
loss_val_log = []
epoch_val_log = []
dice_scores_train_log = []
dice_scores_val_log = []
num_epochs=30
nber_of_tr_samples = len(dataloader_train.dataset)
nber_of_val_samples = len(dataloader_val.dataset)
    
print('START TRAINING...')
for epoch in range(1, num_epochs + 1):
    f1_scores_train = np.zeros(3)
    f1_scores_val = np.zeros(3)

    # Training
    for batch_idx, batch_samples in enumerate(dataloader_train):
        img, seg = batch_samples['img'].to(device), batch_samples['seg'].to(device)
        optimizer.zero_grad()
        prd = model(img)
        loss = F.cross_entropy(prd, seg.squeeze(1))
        loss.backward()
        optimizer.step()

        # Compute the Dice scores for each sample
        seg = seg.squeeze(1) # now size (N,64,64,64), before (N,1,64,64,64)
        prd = torch.argmax(prd, dim=1) # now size (N,64,64,64), before (N,4,64,64,64)
        seg = seg.view(seg.shape[0],-1).cpu().detach().numpy()
        prd = prd.view(prd.shape[0],-1).cpu().detach().numpy()
        for i in range(seg.shape[0]):
          prediction = prd[i,:]
          segmentation = seg[i,:]
          f1_scores = f1_score(segmentation, prediction, labels=[1,2,3], average=None)
          f1_scores_train += f1_scores/nber_of_tr_samples

    dice_scores_train_log.append(np.mean(f1_scores_train))
    loss_train_log.append(loss.item())

    print('+ TRAINING \tEpoch: {} \tLoss: {:.6f}'.format(epoch, loss.item()),
          f'\tClass f1_scores: {np.around(f1_scores_train,2)}, \tAverage f1_scores: {np.around(np.mean(f1_scores_train), 2)}')
    
    # Validation
    if epoch == 1 or epoch % val_interval == 0:
        loss_val = 0
        sum_pts = 0
        with torch.no_grad():
            for data_sample in dataloader_val:
                img, seg = data_sample['img'].to(device), data_sample['seg'].to(device)
                prd = model(img)
                loss_val += F.cross_entropy(prd, seg.squeeze(1), reduction='sum').item()
                sum_pts += np.prod(img_size)

                # Compute the Dice scores for each sample
                seg = seg.squeeze(1) # now size (N,64,64,64), before (N,1,64,64,64)
                prd_prime = torch.argmax(prd, dim=1) # now size (N,64,64,64), before (N,4,64,64,64)
                seg = seg.view(seg.shape[0],-1).cpu().detach().numpy()
                prd_prime = prd_prime.view(prd_prime.shape[0],-1).cpu().detach().numpy()
                for i in range(seg.shape[0]):
                  prediction = prd_prime[i,:]
                  segmentation = seg[i,:]
                  f1_scores = f1_score(segmentation, prediction, labels=[1,2,3], average=None)
                  f1_scores_val += f1_scores/nber_of_val_samples
                
        dice_scores_val_log.append(np.mean(f1_scores_val))
        prd = torch.argmax(prd, dim=1)
        prediction = sitk.GetImageFromArray(prd.cpu().squeeze().numpy().astype(np.uint8))
        

        loss_val /= sum_pts

        loss_val_log.append(loss_val)
        epoch_val_log.append(epoch)

        print('--------------------------------------------------')
        print('+ VALIDATE \tEpoch: {} \tLoss: {:.6f}'.format(epoch, loss_val), 
        f'\tClass f1_scores: {np.around(f1_scores_val,2)}, \tAverage f1_scores: {np.around(np.mean(f1_scores_val), 2)}')
        display_image(sitk.LabelToRGB(prediction))
        print('--------------------------------------------------')

torch.save(model.state_dict(), os.path.join(model_dir, 'model.pt'))

print('\nFinished TRAINING.')

fig, ax = plt.subplots(1, 2, figsize=(15,7))

ax[0].plot(range(1, num_epochs + 1), loss_train_log, c='r', label='train')
ax[0].plot(epoch_val_log, loss_val_log, c='b', label='val')
ax[0].legend(loc='upper right')
ax[0].set_xlabel('epoch')
ax[0].set_ylabel('loss')

ax[1].plot(range(1, num_epochs + 1), dice_scores_train_log, c='r', label='train')
ax[1].plot(epoch_val_log, dice_scores_val_log, c='b', label='val')
ax[1].legend(loc='upper right')
ax[1].set_xlabel('epoch')
ax[1].set_ylabel('average dice score')

plt.show()
```

### Loading the data for a 2 fold cross validation

```python
class AgeRegressionDataset(Dataset):
    """Dataset for image segmentation."""

    def __init__(self, file_list_img, file_list_seg, file_list_msk, img_spacing, img_size, ages):
        self.samples = []
        self.img_names = []
        self.seg_names = []


        for idx, _ in enumerate(tqdm(range(len(file_list_img)), desc='Loading Data')):
            img_path = file_list_img[idx]
            seg_path = file_list_seg[idx]
            msk_path = file_list_msk[idx]

            img = sitk.ReadImage(img_path, sitk.sitkFloat32)

            seg = sitk.ReadImage(seg_path, sitk.sitkInt64)

            msk = sitk.ReadImage(msk_path, sitk.sitkUInt8)

            age = ages[idx]

            #pre=processing
            img = zero_mean_unit_var(img, msk)
            img = resample_image(img, img_spacing, img_size, is_label=False)
            seg = resample_image(seg, img_spacing, img_size, is_label=True)
            msk = resample_image(msk, img_spacing, img_size, is_label=True)

            sample = {'img': img, 'seg': seg, 'msk': msk, 'age': age}

            self.samples.append(sample)
            self.img_names.append(os.path.basename(img_path))
            self.seg_names.append(os.path.basename(seg_path))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, item):
        sample = self.samples[item]

        image = torch.from_numpy(sitk.GetArrayFromImage(sample['img'])).unsqueeze(0)
        seg = torch.from_numpy(sitk.GetArrayFromImage(sample['seg'])).unsqueeze(0)
        msk = torch.from_numpy(sitk.GetArrayFromImage(sample['msk'])).unsqueeze(0)
        age = torch.from_numpy(np.array(sample['age'], dtype=np.float32))

        return {'img': image, 'seg': seg, 'msk': msk, 'age': age}

    def get_sample(self, item):
        return self.samples[item]

    def get_img_name(self, item):
        return self.img_names[item]

    def get_seg_name(self, item):
        return self.seg_names[item]
```

```python
from sklearn.model_selection import train_test_split

# Cross validation datasets
meta_data_reg_train = pd.read_csv(data_dir + 'meta/meta_data_regression_train.csv')
ages = meta_data_reg_train['age']
mean, std = np.mean(ages), np.sqrt(np.var(ages))
ages = (ages - mean)/std
ids_seg_train1, ids_seg_train2, ages1, ages2 = train_test_split(list(meta_data_reg_train['subject_id']),
                                                 ages,
                                                 test_size=0.5,
                                                 shuffle=True)
ages1 = np.array(ages1)
ages2 = np.array(ages2)

# fold1
files_seg_img_train1 = [data_dir + 'images/sub-' + f + '_T1w_unbiased.nii.gz' for f in ids_seg_train1]
files_seg_seg_train1 = [data_dir + 'segs_refs/sub-' + f + '_T1w_seg.nii.gz' for f in ids_seg_train1]
files_seg_msk_train1 = [data_dir + 'masks/sub-' + f + '_T1w_brain_mask.nii.gz' for f in ids_seg_train1]

dataset_train1 = AgeRegressionDataset(files_seg_img_train1, files_seg_seg_train1, files_seg_msk_train1,
                                     img_spacing, img_size, ages1)
dataloader_train1 = torch.utils.data.DataLoader(dataset_train1, batch_size=batch_size, shuffle=True)

# fold2
files_seg_img_train2 = [data_dir + 'images/sub-' + f + '_T1w_unbiased.nii.gz' for f in ids_seg_train2]
files_seg_seg_train2 = [data_dir + 'segs_refs/sub-' + f + '_T1w_seg.nii.gz' for f in ids_seg_train2]
files_seg_msk_train2 = [data_dir + 'masks/sub-' + f + '_T1w_brain_mask.nii.gz' for f in ids_seg_train2]

dataset_train2 = AgeRegressionDataset(files_seg_img_train2, files_seg_seg_train2, files_seg_msk_train2,
                                     img_spacing, img_size, ages2)
dataloader_train2 = torch.utils.data.DataLoader(dataset_train2, batch_size=batch_size, shuffle=True)
```

### The Model

```python
class FCLeNet2(nn.Module):
 
  def __init__(self):
    super(FCLeNet2, self).__init__()

    f = 8 # in features
    k = 5 #kernel size
    s = 2 #stride
    p =2 #padding
   
    self.main = nn.Sequential(
      nn.Conv3d(1, 16, kernel_size=3, stride=1, bias=True, padding=1), #96 -> 48
      nn.LeakyReLU(inplace=True), 
      nn.MaxPool3d(2,2),

      nn.Conv3d(16, 32, kernel_size=k, stride=s, bias=True, padding=p), #48 -> 24
      nn.BatchNorm3d(32),
      nn.LeakyReLU(inplace=True),
      
      nn.Conv3d(32, 64, kernel_size=k,stride=s, bias=True, padding=p), #24 -> 12
      nn.BatchNorm3d(64),
      nn.Dropout3d(0.5),
      nn.LeakyReLU(inplace=True),

      nn.Conv3d(64, 128, kernel_size=k, stride=s, bias=True, padding=p), #12 -> 6
      nn.BatchNorm3d(128), 
      # nn.Dropout3d(0.5),
      nn.LeakyReLU(inplace=True),
      
      nn.Conv3d(128, 128, kernel_size=k, stride=s, bias=True, padding=p), #6->3
      nn.BatchNorm3d(128),
      nn.Dropout3d(0.5),
      nn.LeakyReLU(inplace=True),

      nn.Conv3d(128, 128, kernel_size=3, stride=1, bias=True, padding=0), #3 -> 1
      nn.BatchNorm3d(128),
      # nn.Dropout3d(0.5),
      nn.LeakyReLU(inplace=True),
    )
    self.fc = nn.Linear(128, 1)
  
  def forward(self, x):
    x = self.main(x)
    x = x.view(x.size(0), -1)
    y_pred = self.fc(x) # y_pred.shape = [N, 1, 1, 1, 1] # adding a linear layer activation

    return y_pred
```

```python
class FCLeNet3(nn.Module):
    def __init__(self):
        super(FCLeNet3, self).__init__()
        
        self.net = nn.Sequential(
            
            nn.Conv3d(1, 8, kernel_size=3, bias=False), 
            nn.BatchNorm3d(8),
            nn.ReLU(),
            nn.MaxPool3d(kernel_size=2),   
            
            nn.Conv3d(8, 32, kernel_size=3, bias=False), 
            nn.ReLU(),
            nn.BatchNorm3d(32),
            nn.MaxPool3d(kernel_size=2),
            
            nn.Conv3d(32, 64, kernel_size=3, bias=False),
            nn.ReLU(),
            nn.BatchNorm3d(64),
            nn.MaxPool3d(kernel_size=2),
            
            nn.Conv3d(64, 128, kernel_size=3, bias=False),
            nn.ReLU(),
            nn.BatchNorm3d(128),
            nn.MaxPool3d(kernel_size=2),

            

            nn.Conv3d(128, 256, kernel_size=2, bias=False),
            nn.ReLU(),
            nn.BatchNorm3d(256),
            nn.MaxPool3d(kernel_size=2),
        )
        

        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(256, 1)


    def forward(self, x):
        
        x = self.net(x)
        x = x.view(x.size(0), -1)
        out = self.fc(x)
        
        return out
```

```python
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)
```

### 2 fold cross validation training loop


```python
num_epochs = 30
val_interval = 2
learning_rate = 0.001
loss_function = nn.MSELoss()

for i in range(2):
  
  # Initialize model
  model = FCLeNet3()
  # model.apply(weights_init)
  params = sum(p.numel() for p in model.parameters() if p.requires_grad)
  print("Total number of parameters: {}".format(params))
  model = model.to(device)
  optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

  # Set training and validation folds
  if i==0:
    print('\n*** Training on fold number 1, validating on fold number 2 ***\n')
    dataloader_train = dataloader_train1
    dataloader_val = dataloader_train2
  else:
    print('\n*** Training on fold number 2, validating on fold number 1 ***\n')
    dataloader_train = dataloader_train2
    dataloader_val = dataloader_train1

  tr_loss_list = []
  tr_mae_list = []
  val_loss_list = []
  val_mae_list = []
  val_epoch_list = []

  for epoch in range(1, num_epochs + 1):
    
    ########
    # Train
    ########
    model.train()
    epoch_loss = 0
    epoch_mae = 0
    for idx, batch in enumerate(dataloader_train):

      # predict and take step
      images, ages = batch['img'], batch['age'].to(dtype=torch.float32)
      images = images.to(device)
      ages = ages.to(device)
      
      optimizer.zero_grad()
      predictions = model(images).squeeze()
      loss = loss_function(predictions, ages)
      loss.backward()
      optimizer.step()

      # Save loss 
      epoch_loss += loss.item()
      unnormalized_ages = (ages*std+mean).detach().cpu().numpy()
      unnormalized_preds = (predictions*std+mean).detach().cpu().numpy()
      epoch_mae += np.mean(np.abs(unnormalized_ages-unnormalized_preds))

    epoch_loss = epoch_loss/(idx+1)
    epoch_mae = epoch_mae/(idx+1)
    print(f'+ TRAINING \tEpoch: {epoch} \tLoss: {epoch_loss} \tMAE: {epoch_mae}')
    tr_loss_list.append(epoch_loss)
    tr_mae_list.append(epoch_mae)

    ###########
    # Validate
    ###########
    if epoch==1 or epoch % val_interval == 0:

      model.eval()
      epoch_loss = 0
      epoch_mae = 0
      for idx, batch in enumerate(dataloader_val):

        # Predict
        images, ages = batch['img'], batch['age'].to(dtype=torch.float32)
        images = images.to(device)
        ages = ages.to(device)   
        predictions = model(images).squeeze()
        loss = loss_function(predictions, ages)
        
        # Save loss
        epoch_loss += loss.item()
        unnormalized_ages = (ages*std+mean).detach().cpu().numpy()
        unnormalized_preds = (predictions*std+mean).detach().cpu().numpy()
        epoch_mae += np.mean(np.abs(unnormalized_ages-unnormalized_preds))
      
      epoch_loss = epoch_loss/(idx+1)
      epoch_mae = epoch_mae/(idx+1)
      print(f'+ VALIDATION \tEpoch: {epoch} \tLoss: {epoch_loss} \tMAE: {epoch_mae}')
      val_loss_list.append(epoch_loss)
      val_mae_list.append(epoch_mae)
      val_epoch_list.append(epoch)

  ##############
  # Plot curves
  ##############
  fig, axes = plt.subplots(1,2, figsize=(13,6))
  ax = axes[0]
  ax.plot(range(1, num_epochs+1), tr_loss_list, 'r', label='train loss')
  ax.plot(val_epoch_list, val_loss_list, 'g', label='val loss')
  ax.set_xlabel('Epochs')
  ax.set_ylabel('Loss')
  ax.legend(loc='upper right')

  ax = axes[1]
  ax.plot(range(1, num_epochs+1), tr_mae_list, 'r', label='train MAE')
  ax.plot(val_epoch_list, val_mae_list, 'g', label='val MAE')
  ax.set_xlabel('Epochs')
  ax.set_ylabel('MAE')
  ax.legend(loc='upper right')
  plt.show()

```

```python
out_dir = './output'
model_dir = os.path.join(out_dir, 'model')
if not os.path.exists(model_dir):
    os.makedirs(model_dir)

num_epochs = 35
val_interval = 2
learning_rate = 0.001
loss_function = nn.MSELoss()

# Initialize model
model = FCLeNet3()
# model.apply(weights_init)
params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("Total number of parameters in FCLeNet2: {}".format(params))
model = model.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)



tr_loss_list = []
tr_mae_list = []
val_loss_list = []
val_mae_list = []
val_epoch_list = []

for epoch in range(1, num_epochs + 1):
  
  ########
  # Train
  ########
  model.train()
  epoch_loss = 0
  epoch_mae = 0
  for idx, batch in enumerate(dataloader_train):

    # predict and take step
    images, ages = batch['img'], batch['age'].to(dtype=torch.float32)
    images = images.to(device)
    ages = ages.to(device)
    
    optimizer.zero_grad()
    predictions = model(images).squeeze()
    loss = loss_function(predictions, ages)
    loss.backward()
    optimizer.step()

    # Save loss 
    epoch_loss += loss.item()
    unnormalized_ages = (ages*std+mean).detach().cpu().numpy()
    unnormalized_preds = (predictions*std+mean).detach().cpu().numpy()
    epoch_mae += np.mean(np.abs(unnormalized_ages-unnormalized_preds))

  epoch_loss = epoch_loss/(idx+1)
  epoch_mae = epoch_mae/(idx+1)
  print(f'+ TRAINING \tEpoch: {epoch} \tLoss: {epoch_loss} \tMAE: {epoch_mae}')
  tr_loss_list.append(epoch_loss)
  tr_mae_list.append(epoch_mae)

  ###########
  # Validate
  ###########
  if epoch==1 or epoch % val_interval == 0:

    model.eval()
    epoch_loss = 0
    epoch_mae = 0
    for idx, batch in enumerate(dataloader_val):

      # Predict
      images, ages = batch['img'], batch['age'].to(dtype=torch.float32)
      images = images.to(device)
      ages = ages.to(device)   
      predictions = model(images).squeeze()
      loss = loss_function(predictions, ages)
      
      # Save loss
      epoch_loss += loss.item()
      unnormalized_ages = (ages*std+mean).detach().cpu().numpy()
      unnormalized_preds = (predictions*std+mean).detach().cpu().numpy()
      epoch_mae += np.mean(np.abs(unnormalized_ages-unnormalized_preds))
    
    epoch_loss = epoch_loss/(idx+1)
    epoch_mae = epoch_mae/(idx+1)
    print(f'+ VALIDATION \tEpoch: {epoch} \tLoss: {epoch_loss} \tMAE: {epoch_mae}')
    val_loss_list.append(epoch_loss)
    val_mae_list.append(epoch_mae)
    val_epoch_list.append(epoch)

##############
# Plot curves
##############
fig, axes = plt.subplots(1,2, figsize=(13,6))
ax = axes[0]
ax.plot(range(1, num_epochs+1), tr_loss_list, 'r', label='train loss')
ax.plot(val_epoch_list, val_loss_list, 'g', label='val loss')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.legend(loc='upper right')

ax = axes[1]
ax.plot(range(1, num_epochs+1), tr_mae_list, 'r', label='train MAE')
ax.plot(val_epoch_list, val_mae_list, 'g', label='val MAE')
ax.set_xlabel('Epochs')
ax.set_ylabel('MAE')
ax.legend(loc='upper right')
plt.show()

torch.save(model.state_dict(), os.path.join(model_dir, 'reg_model.pt'))
```
